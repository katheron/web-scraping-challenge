````python
# Import BeautifulSoup
from bs4 import BeautifulSoup as bs
# Import Pandas
import pandas as pd
import requests
````

````python
# Import Splinter and set the chromedriver path
from splinter import Browser
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
````

Scrape the NASA site and collect the latest News Title and Paragraph Text. Assign the text to variables that you can reference later.

````python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
print(soup.prettify())
````

````python
# Collect the Latest News Title and Paragraph Text. Assign the text to variables that you can reference later

news_title = soup.find("div", class_='content_title').text
print(f"Latest News Title:{news_title}")

news_p = soup.find("div", class_='rollover_description_inner').text
print(f"Paragraph Text: {news_p}")
````

Scrape the JPL Featured Space Image

Use splinter to navigate the site and find the image url for the current Featured Mars Image and assign the url string to a variable called `featured_image_url`.

````python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
````

````python
html = browser.html
soup = bs(html, "html.parser")
````

````python
img_source = soup.find(class_ = "carousel_item")['style']
print(img_source)
````

````python
img_string = img_source.split("'")[1]
print(img_string)
````

````python
base_url = "https://jpl.nasa.gov"

featured_image_url = base_url + img_string
print(featured_image_url)
````
